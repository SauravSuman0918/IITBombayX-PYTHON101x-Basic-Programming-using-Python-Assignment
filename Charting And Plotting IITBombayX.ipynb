# Least square fit

```python
# we need to fit a line through points for the equation t^2=m.L+c
#In matrix it can be represented as Tsq=A.p
we need to find p to plot the line
```

```python
L,t=np.loadtxt('pendulum.txt',unpack=True,delimiter=";")
```

```python
tsq=t*t
```

```python
A=np.array([L,np.ones_like(L)])
A=A.T  #transpose
```

```python
result=np.linalg.lstsq(A,tsq)
coef=result[0]  # coefficient
```

```python
Tline=coef[0]*L +coef[1]
Tline.shape
```

```python
#we have to plot Tline vs L to get the least squares fit line
plt.plot(L,Tline,'r')
```
